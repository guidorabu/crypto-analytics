# 1. Importación de Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Configuración de gráficos

```python
sns.set(style="whitegrid")
plt.style.use("ggplot")
```

# 2. Carga de Datos

```python
data_path = "../data/processed/cryptos_processed.csv"
df = pd.read_csv(data_path)
```

# 3. Inspección Inicial de Datos

```python
print("Primeras filas del dataset:")
display(df.head())

print("\nInformación del dataset:")
print(df.info())

print("\nEstadísticas descriptivas:")
display(df.describe())
```

# 4. Visualizaciones Básicas


## Gráfico de barras - Top 10 precios actuales

```python
top_prices = df.nlargest(10, 'current_price')
plt.figure(figsize=(10, 6))
sns.barplot(x='current_price', y='name', data=top_prices, palette='viridis')
plt.title("Top 10 Criptomonedas por Precio Actual")
plt.xlabel("Precio Actual (USD)")
plt.ylabel("Criptomoneda")
plt.show()
```


## Gráfico de barras - Top 10 por Market Cap

```python
top_market_cap = df.nlargest(10, 'market_cap')
plt.figure(figsize=(10, 6))
sns.barplot(x='market_cap', y='name', data=top_market_cap, palette='magma')
plt.title("Top 10 Criptomonedas por Market Cap")
plt.xlabel("Market Cap (USD)")
plt.ylabel("Criptomoneda")
plt.show()
```

## Histograma del precio actual

```python
df['current_price'] = df['current_price'].replace([np.inf, -np.inf], np.nan).dropna()

# Graficar el histograma del precio actual
plt.figure(figsize=(10, 8))
plt.hist(df['current_price'], bins=20, edgecolor='black')
plt.yscale('log')  # Escala logarítmica en el eje y
plt.xlabel('Precio Actual (USD)')
plt.ylabel('Frecuencia')
plt.title('Distribución de los Precios Actuales de Criptomonedas')
plt.show()
```

```python
# Filtrar criptomonedas con precios mayores a un umbral (por ejemplo, 1000 USD)
df_high_price = df[df['current_price'] > 1000]
df_low_price = df[df['current_price'] <= 1000]

# Histograma para criptos con precios bajos
plt.figure(figsize=(10, 8))
plt.hist(df_low_price['current_price'], bins=20, edgecolor='black')
plt.xlabel('Precio Actual (USD)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios de Criptomonedas con Precios Bajos')
plt.show()

# Histograma para criptos con precios altos
plt.figure(figsize=(10, 8))
plt.hist(df_high_price['current_price'], bins=20, edgecolor='black')
plt.xlabel('Precio Actual (USD)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios de Criptomonedas con Precios Altos')
plt.show()
```

# 5. Análisis de Correlaciones

```python
# Filtrar sólo las columnas numéricas
numerical_df = df.select_dtypes(include=[np.number])

# Calcular y graficar la matriz de correlación
plt.figure(figsize=(12, 8))
correlation_matrix = numerical_df.corr()

# Mapa de calor de la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación entre Variables Numéricas")
plt.show()
```
